```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from openpyxl import Workbook
```

### Set Path for webdriver

```python

options = webdriver.ChromeOptions()
options.add_argument('--no-sandbox')
options.add_argument('--headless')
options.add_argument('--disable-gpu')
options.add_argument('--disable-dve-shm-uage')

browser = webdriver.Chrome(options = options)
```

### For Accessing Title of Blog


```python
browser.get('https://rategain.com/blog')
```

```python
for element in browser.find_elements(By.XPATH, '//div[@class="content"]/h6/a'):
  print(element.text)
```

### For Accessing Date of Blog


```python
for element in browser.find_elements(By.XPATH, '//div[@class="blog-detail"]/div[@class="bd-item"][1]/span'):
  print(element.text)
```

### For Acesssing Image of Blog

```python
for element in browser.find_elements(By.XPATH, '//div[@class="img"]/a'):
  print(element.get_attribute('data-bg'))
```

### For Accessing Likes of Blog


```python
for element in browser.find_elements(By.XPATH, '//a[@class="zilla-likes"]/span'):
  print(element.text)
```

### To Find Total Count Of Pages

```python
element = browser.find_element(By.XPATH, '//a[@class="page-numbers"][last()]')
print(element.text)
c=element.text
```

### **FINAL CODE**

```python
#result array
res=[]

#find the number of pages
element = browser.find_element(By.XPATH, '//a[@class="page-numbers"][last()]')
count=int(element.text)

#default path for each page
path='https://rategain.com/blog/page/'

#counter for blogs
a=0

#traverse over all the pages
for i in range(count):
  browser.get(path + str(i+1))

  #count the number of articles on the page
  v=browser.find_elements(By.XPATH, '//article')
  n=len(v)

  #traverse over all the articles
  for j in range(n-1):
    res.append([])

    #title of article
    title = browser.find_element(By.XPATH, '//article['+str(j+1)+']//div[@class="content"]/h6/a')
    res[a].append(title.text)

    #date of article
    date = browser.find_element(By.XPATH, '//article['+str(j+1)+']//div[@class="blog-detail"]/div[@class="bd-item"][1]/span')
    res[a].append(date.text)

    #to get the class of the article
    dex=browser.find_element(By.XPATH, '//article//article['+str(j+1)+']')
    cls = dex.get_attribute('class')
    lis=list(cls.split(" "))

    #image of the article
    if "with-image" in lis:
      img=browser.find_element(By.XPATH, '//article['+str(j+1)+']//div[@class="img"]/a')
      res[a].append(img.get_attribute('data-bg'))
    else:
      res[a].append("No image found")

    #likes of the article
    like = browser.find_element(By.XPATH, '//article['+str(j+1)+']//a[@class="zilla-likes"]/span')
    res[a].append(like.text)

    a+=1


```

### **FINAL ANSWER**

```python
for i in range(len(res)):
  print(res[i])
```

### Saving data to '.xlsx' workbook

```python
from openpyxl import Workbook

filename = "blog_data.xlsx"

#create a new empty workbook.
workbook = Workbook()
sheet = workbook.active

sheet["A1"]= "Title"
sheet["B1"]= "Date"
sheet["C1"]= "Image URL"
sheet["D1"]= "Likes Count"


for i in range(len(res)):
  sheet["A"+str(i+2)]=res[i][0]
  sheet["B"+str(i+2)]=res[i][1]
  sheet["C"+str(i+2)]=res[i][2]
  sheet["D"+str(i+2)]=res[i][3]

#save the spreadsheet
workbook.save(filename=filename)
```
